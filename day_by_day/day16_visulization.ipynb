```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]   # X-axis values
y = [2, 4, 6, 8, 10]  # Y-axis values

# Create line graph
plt.plot(x, y, color='blue', marker='o', linestyle='-')

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Line Graph")

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D']
values1 = [4, 7, 1, 8]
values2 = [2, 4, 5, 3]

# Create stacked bar chart
plt.bar(categories, values1, color='skyblue', label='Group 1')
plt.bar(categories, values2, bottom=values1, color='orange', label='Group 2')

# Labels and title
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Stacked Bar Chart")
plt.legend()

# Show chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

# Horizontal bar chart
plt.barh(categories, values, color='lightcoral', edgecolor='black')

# Labels and title
plt.xlabel("Values")      # now x-axis = values
plt.ylabel("Categories")  # y-axis = categories
plt.title("Horizontal Bar Chart")

# Show chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample categories
categories = ['A', 'B', 'C', 'D']

# 5 sets of values
v1 = [3, 5, 2, 6]
v2 = [4, 2, 3, 5]
v3 = [2, 3, 4, 1]
v4 = [1, 2, 2, 3]
v5 = [2, 1, 3, 2]

# Plot stacked bars
plt.bar(categories, v1, label='Value 1')
plt.bar(categories, v2, bottom=v1, label='Value 2')
plt.bar(categories, v3, bottom=[i+j for i,j in zip(v1, v2)], label='Value 3')
plt.bar(categories, v4, bottom=[i+j+k for i,j,k in zip(v1, v2, v3)], label='Value 4')
plt.bar(categories, v5, bottom=[i+j+k+l for i,j,k,l in zip(v1, v2, v3, v4)], label='Value 5')

# Labels and title
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Stacked Bar Chart with 5 Values")
plt.legend()

# Show chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
labels = ['A', 'B', 'C', 'D']
values = [30, 25, 20, 25]

# Create pie chart
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)

# Title
plt.title("Basic Pie Chart")

# Show chart
plt.show()
```
