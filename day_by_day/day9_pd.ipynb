
üêº Pandas Overview

Library Name: pandas

Installation:

pip install pandas

Purpose: Pandas is a Python library used for data analysis and data manipulation.

Main Data Structures:

Series ‚Üí 1D (like a single column)

DataFrame ‚Üí 2D (like an Excel sheet or SQL table)




---

üìÇ File Types Pandas Can Read

Pandas is very powerful ‚Äî it can read many file types directly.
Here are the most common ones with examples:


---

1. CSV Files

import pandas as pd

df = pd.read_csv("data.csv")
print(df.head())

‚úÖ Use When: You have comma-separated tabular data.
üìå Function: pd.read_csv()


---

2. Excel Files (.xlsx, .xls)

import pandas as pd

df = pd.read_excel("data.xlsx", sheet_name="Sheet1")
print(df.head())

‚úÖ Use When: You have spreadsheet data with multiple sheets.
üìå Function: pd.read_excel()
üîë Extra Requirement: openpyxl library (install with pip install openpyxl)


---

3. JSON Files

import pandas as pd

df = pd.read_json("data.json")
print(df.head())

‚úÖ Use When: You have structured JSON data.
üìå Function: pd.read_json()


---

4. Parquet Files

import pandas as pd

df = pd.read_parquet("data.parquet")
print(df.head())

‚úÖ Use When: You work with large columnar datasets (very fast).
üìå Function: pd.read_parquet()
üîë Extra Requirement: pyarrow or fastparquet


---

5. HTML Tables

import pandas as pd

tables = pd.read_html("https://example.com/tablepage")
print(tables[0])

‚úÖ Use When: You want to extract tables from websites.
üìå Function: pd.read_html()
üîë Extra Requirement: lxml or html5lib


---

6. SQL Databases

import pandas as pd
import sqlite3

conn = sqlite3.connect("mydatabase.db")
df = pd.read_sql("SELECT * FROM users", conn)
print(df.head())

‚úÖ Use When: You want to read directly from a database.
üìå Function: pd.read_sql()


---

7. Other Formats

HDF5: pd.read_hdf()

Feather: pd.read_feather()

ORC: pd.read_orc()

Pickle: pd.read_pickle()

Clipboard: pd.read_clipboard() (reads data copied to clipboard)



---

üîë Summary Table

File Type	Pandas Function	Extra Library?

CSV	pd.read_csv()	No
Excel (.xlsx/.xls)	pd.read_excel()	‚úÖ openpyxl
JSON	pd.read_json()	No
Parquet	pd.read_parquet()	‚úÖ pyarrow or fastparquet
HTML	pd.read_html()	‚úÖ lxml or html5lib
SQL	pd.read_sql()	No (needs DB connection)
HDF5	pd.read_hdf()	‚úÖ tables
Feather	pd.read_feather()	‚úÖ pyarrow
ORC	pd.read_orc()	‚úÖ pyarrow
Pickle	pd.read_pickle()	No



---

üí° Key Points to Remember

pandas itself is the main library ‚Äî you just import pandas as pd.

You don‚Äôt need to write loops ‚Äî pandas loads data into a DataFrame, which is easy to analyze.

Some formats (Excel, Parquet, ORC) need extra libraries installed.
